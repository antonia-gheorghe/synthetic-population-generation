```python
import json
import numpy as np
import plotly.graph_objects as go
import plotly.io as pio

# Step 1: Load the JSON file
with open('counts.json') as f:
    data = json.load(f)

# Step 2: Extract age counts from ground truth (use relevant keys from your data)
age_groups = ['0_4-', '5_7-', '8_9-', '10_14-', '15-', '16_17-', '18_19-', '20_24-', '25_29-', '30_34-', '35_39-', '40_44-', '45_49-', '50_54-', '55_59-', '60_64-', '65_69-', '70_74-', '75_79-', '80_84-', '85+-']

# Define more readable age labels
age_labels = ['0-4', '5-7', '8-9', '10-14', '15', '16-17', '18-19', '20-24', '25-29', '30-34', '35-39', '40-44', '45-49', '50-54', '55-59', '60-64', '65-69', '70-74', '75-79', '80-84', '85+']

ground_truth_age_counts = {
    age: sum([data['age_sex_ethnicity_counts'][key] for key in data['age_sex_ethnicity_counts'] if age in key])
    for age in age_groups
}

# Step 3: Extract age counts from results
results_age_counts = {
    age: sum([data['age_sex_ethnicity_religion_marital_counts'][key] for key in data['age_sex_ethnicity_religion_marital_counts'] if age in key])
    for age in age_groups
}

# Step 4: Prepare categories (age groups) and values (counts)
# Use age_labels instead of age_groups for better display
categories = age_labels
ground_truth_values = list(ground_truth_age_counts.values())
results_values = list(results_age_counts.values())

print(ground_truth_values)
print(results_values)

# Step 5: Apply log transformation (log base 2)
# Replace zeros with a small value (e.g., 1) to avoid log(0) issues
small_value = 1  # A small value to replace zeros
log_ground_truth_values = [np.log2(x + small_value) for x in ground_truth_values]
log_results_values = [np.log2(x + small_value) for x in results_values]

# Step 6: Create radar chart (spider chart) using Plotly
fig = go.Figure()

# Add trace for the ground truth (log scale)
fig.add_trace(go.Scatterpolar(
    r=log_ground_truth_values,
    theta=categories,  # Use age_labels for better readability
    fill='toself',
    name='Ground Truth (Log Scale)'
))

# Add trace for the results (log scale)
fig.add_trace(go.Scatterpolar(
    r=log_results_values,
    theta=categories,  # Use age_labels for better readability
    fill='toself',
    name='Results (Log Scale)'
))

# Step 7: Update layout for better visuals
fig.update_layout(
    polar=dict(
        radialaxis=dict(
            visible=True,
            range=[0, max(max(log_ground_truth_values), max(log_results_values))]  # Adjust range based on the maximum log value
        )),
    showlegend=False,
    # title='Comparison of Age Counts (Ground Truth vs Results) - Logarithmic Scale'
)

# Step 8: Show the radar chart
fig.show()

# # Save the figure as a PDF
# pio.write_image(fig, 'radar_chart_age.pdf', format='pdf')
```

```python
import json
import plotly.graph_objects as go
import plotly.io as pio

# Step 1: Load the JSON file
with open('acc_data.json') as f:
    accuracy_data = json.load(f)

# Step 2: Extract the accuracies for each category (sex, age, ethnicity, religion, marital) across iterations
iterations = list(accuracy_data.keys())

# Initialize lists for different accuracies
sex_acc = []
age_acc = []
ethnicity_acc = []
religion_acc = []
marital_acc = []

freq = 10

# Step 3: Loop through each iteration and extract accuracies for every 10 epochs
for iteration in iterations:
    if int(iteration) % freq == 0:  # Only include every 10th iteration
        # Extract accuracies for 'sex, age, ethnicity'
        sex_acc.append(accuracy_data[iteration]["sex, age, ethnicity"]["sex"])
        age_acc.append(accuracy_data[iteration]["sex, age, ethnicity"]["age"])
        ethnicity_acc.append(accuracy_data[iteration]["sex, age, ethnicity"]["ethnicity"])

        # Extract accuracies for 'sex, age, religion'
        religion_acc.append(accuracy_data[iteration]["sex, age, religion"]["religion"])

        # Extract accuracies for 'sex, age, marital'
        marital_acc.append(accuracy_data[iteration]["sex, age, marital"]["marital"])

# Filter iterations to include only multiples of 10
filtered_iterations = [int(iteration) // freq for iteration in iterations if int(iteration) % freq == 0]

# Step 4: Create the plot using Plotly with different shades of blue
fig = go.Figure()

# Add trace for Sex accuracy (Light Blue)
fig.add_trace(go.Scatter(
    x=filtered_iterations,
    y=sex_acc,
    mode='lines',
    name='Sex',
    line=dict(color='#AEC7E8'),  # Light Blue color
    marker=dict(symbol='circle', size=6)
))

# Add trace for Age accuracy (Sky Blue)
fig.add_trace(go.Scatter(
    x=filtered_iterations,
    y=age_acc,
    mode='lines',
    name='Age',
    line=dict(color='#1F77B4'),  # Sky Blue color
    marker=dict(symbol='square', size=6)
))

# Add trace for Ethnicity accuracy (Steel Blue)
fig.add_trace(go.Scatter(
    x=filtered_iterations,
    y=ethnicity_acc,
    mode='lines',
    name='Ethnicity',
    line=dict(color='#7FB3D5'),  # Steel Blue color
    marker=dict(symbol='triangle-up', size=6)
))

# Add trace for Religion accuracy (Dark Blue)
fig.add_trace(go.Scatter(
    x=filtered_iterations,
    y=religion_acc,
    mode='lines',
    name='Religion',
    line=dict(color='#21618C'),  # Dark Blue color
    marker=dict(symbol='triangle-down', size=6)
))

# Add trace for Marital accuracy (Midnight Blue)
fig.add_trace(go.Scatter(
    x=filtered_iterations,
    y=marital_acc,
    mode='lines',
    name='Marital',
    line=dict(color='#2C3E50'),  # Midnight Blue color
    marker=dict(symbol='diamond', size=6)
))

# Step 5: Customize the plot

# Only show labels for multiples of 500 epochs
tickvals_500 = [i for i in filtered_iterations if i % 50 == 0]  # Since each tick is 10 epochs, 50 * 10 = 500 epochs
ticktext_500 = [str(i * 10) for i in tickvals_500]  # Display the actual epoch value (multiples of 500)

fig.update_layout(
    xaxis_title='Iterations',
    yaxis_title='Accuracy',
    template='plotly_white',
    font=dict(size=14),
    width=1200,
    height=800,
    legend=dict(
        x=1,  # Position legend on the right
        y=0.1,  # Position legend at the bottom
        xanchor='right',  # Align the legend box with the right edge
        yanchor='bottom',  # Align the legend box with the bottom edge
        bgcolor='rgba(255, 255, 255, 0.5)',  # Optional: add a semi-transparent background
        bordercolor='black',  # Optional: add a border
        borderwidth=1,  # Border width
        font=dict(size=18)
    ),
    xaxis=dict(
        tickmode='array',
        tickvals=tickvals_500,
        ticktext=ticktext_500  # Display iterations as multiples of 500
    )
)

# Step 6: Display the plot
fig.show()

# Step 7: Save the figure as a PDF
pio.write_image(fig, 'accuracy_household_attributes_500_epochs.pdf', format='pdf')
```

```python
import json
import plotly.graph_objects as go
import plotly.io as pio

# Step 1: Load the JSON file
with open('acc_data.json') as f:
    accuracy_data = json.load(f)

# Step 2: Extract the net accuracies for each category (ethnicity, marital, religion) across iterations
iterations = list(accuracy_data.keys())

# Initialize lists for net accuracies
net_sex_age_ethnicity = []
net_sex_age_marital = []
net_sex_age_religion = []

# Step 3: Loop through each iteration and extract net accuracies for every 10 epochs
actual_epochs = []  # Store the actual epoch numbers for the x-axis
for iteration in iterations:
    if int(iteration) % 10 == 0:  # Only include every 10th iteration
        actual_epochs.append(int(iteration))  # Store the actual epoch number
        # Extract net accuracies for 'sex, age, ethnicity'
        net_sex_age_ethnicity.append(accuracy_data[iteration]["sex, age, ethnicity"]["net"])

        # Extract net accuracies for 'sex, age, marital'
        net_sex_age_marital.append(accuracy_data[iteration]["sex, age, marital"]["net"])

        # Extract net accuracies for 'sex, age, religion'
        net_sex_age_religion.append(accuracy_data[iteration]["sex, age, religion"]["net"])


# Print final values for debugging
print('net_sex_age_ethnicity', net_sex_age_ethnicity[-1])
print('net_sex_age_marital', net_sex_age_marital[-1])
print('net_sex_age_religion', net_sex_age_religion[-1])

# Step 4: Create the plot using Plotly
fig = go.Figure()

# Add trace for 'sex, age, ethnicity' net accuracy with Navy Blue
fig.add_trace(go.Scatter(
    x=actual_epochs,
    y=net_sex_age_ethnicity,
    mode='lines',
    name='Sex, Age, Ethnicity (Net)',
    line=dict(color='#000080'),  # Navy Blue
    marker=dict(symbol='circle', size=6)
))

# Add trace for 'sex, age, marital' net accuracy with Light Blue
fig.add_trace(go.Scatter(
    x=actual_epochs,
    y=net_sex_age_marital,
    mode='lines',
    name='Sex, Age, Marital (Net)',
    line=dict(color='#ADD8E6'),  # Light Blue
    marker=dict(symbol='diamond', size=6)
))

# Add trace for 'sex, age, religion' net accuracy with Royal Blue
fig.add_trace(go.Scatter(
    x=actual_epochs,
    y=net_sex_age_religion,
    mode='lines',
    name='Sex, Age, Religion (Net)',
    line=dict(color='#4169E1'),  # Royal Blue
    marker=dict(symbol='triangle-up', size=6)
))

# Step 5: Customize the plot
fig.update_layout(
    xaxis_title='Epochs',
    yaxis_title='Net Accuracy',
    template='plotly_white',
    font=dict(size=14),
    width=1200,
    height=800,
    legend=dict(
        x=1,  # Position legend on the right
        y=0.1,  # Position legend at the bottom
        xanchor='right',  # Align the legend box with the right edge
        yanchor='bottom',  # Align the legend box with the bottom edge
        bgcolor='rgba(255, 255, 255, 0.5)',  # Optional: add a semi-transparent background
        bordercolor='black',  # Optional: add a border
        borderwidth=1,  # Border width
        font=dict(size=18)
    ),
)

# Step 6: Display the plot
fig.show()

# Step 7: Save the figure as a PDF (optional)
pio.write_image(fig, 'net_accuracy_10_epochs.pdf', format='pdf')
```
