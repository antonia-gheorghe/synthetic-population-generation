```python
import plotly.io as pio
import numpy as np
import json

# Step 1: Load the JSON file
with open('counts.json') as f:
    data = json.load(f)

# Step 2: Extract religion counts from ground truth (use relevant keys from your data)
ground_truth_religion_counts = {
    'Christian': sum([data['hh_religion_counts'][key] for key in data['hh_religion_counts'] if '-C' in key]),
    'Buddhist': sum([data['hh_religion_counts'][key] for key in data['hh_religion_counts'] if '-B' in key]),
    'Hindu': sum([data['hh_religion_counts'][key] for key in data['hh_religion_counts'] if '-H' in key]),
    'Jewish': sum([data['hh_religion_counts'][key] for key in data['hh_religion_counts'] if '-J' in key]),
    'Muslim': sum([data['hh_religion_counts'][key] for key in data['hh_religion_counts'] if '-M' in key]),
    'Sikh': sum([data['hh_religion_counts'][key] for key in data['hh_religion_counts'] if '-S' in key]),
    'Other Religion': sum([data['hh_religion_counts'][key] for key in data['hh_religion_counts'] if '-OR' in key]),
    'No Religion': sum([data['hh_religion_counts'][key] for key in data['hh_religion_counts'] if '-NR' in key]),
    'Not Stated': sum([data['hh_religion_counts'][key] for key in data['hh_religion_counts'] if '-NS' in key])
}

# Step 3: Extract religion counts from results
results_religion_counts = {
    'Christian': sum([data['hh_ethnicity_religion_counts'][key] for key in data['hh_ethnicity_religion_counts'] if '-C' in key]),
    'Buddhist': sum([data['hh_ethnicity_religion_counts'][key] for key in data['hh_ethnicity_religion_counts'] if '-B' in key]),
    'Hindu': sum([data['hh_ethnicity_religion_counts'][key] for key in data['hh_ethnicity_religion_counts'] if '-H' in key]),
    'Jewish': sum([data['hh_ethnicity_religion_counts'][key] for key in data['hh_ethnicity_religion_counts'] if '-J' in key]),
    'Muslim': sum([data['hh_ethnicity_religion_counts'][key] for key in data['hh_ethnicity_religion_counts'] if '-M' in key]),
    'Sikh': sum([data['hh_ethnicity_religion_counts'][key] for key in data['hh_ethnicity_religion_counts'] if '-S' in key]),
    'Other Religion': sum([data['hh_ethnicity_religion_counts'][key] for key in data['hh_ethnicity_religion_counts'] if '-OR' in key]),
    'No Religion': sum([data['hh_ethnicity_religion_counts'][key] for key in data['hh_ethnicity_religion_counts'] if '-NR' in key]),
    'Not Stated': sum([data['hh_ethnicity_religion_counts'][key] for key in data['hh_ethnicity_religion_counts'] if '-NS' in key])
}

# Step 4: Prepare categories (religions) and values (counts)
categories = list(ground_truth_religion_counts.keys())
ground_truth_values = list(ground_truth_religion_counts.values())
results_values = list(results_religion_counts.values())

# Step 5: Apply log transformation (log base 10)
# Replace zeros with a small value (e.g., 1) to avoid log(0) issues
small_value = 1  # A small value to replace zeros

log_ground_truth_values = [np.log2(x + small_value) for x in ground_truth_values]
log_results_values = [np.log2(x + small_value) for x in results_values]

# Step 6: Create radar chart (spider chart) using Plotly
fig = go.Figure()

# Add trace for the ground truth (log scale)
fig.add_trace(go.Scatterpolar(
    r=log_ground_truth_values,
    theta=categories,
    fill='toself',
    name='Ground Truth (Log Scale)'
))

# Add trace for the results (log scale)
fig.add_trace(go.Scatterpolar(
    r=log_results_values,
    theta=categories,
    fill='toself',
    name='Results (Log Scale)'
))

# Step 7: Update layout for better visuals
fig.update_layout(
    # title='Comparison of Religion Counts (Ground Truth vs Results) - Logarithmic Scale',
    polar=dict(
        radialaxis=dict(
            visible=True,
            range=[0, max(max(log_ground_truth_values), max(log_results_values))]  # Adjust range based on the maximum log value
        )),
    showlegend=False
)

# Step 8: Show the radar chart
fig.show()

# Save the figure as a PDF
pio.write_image(fig, 'radar_chart_religion.pdf', format='pdf')
```

```python
# Step 1: Load the JSON file
with open('loss_data.json') as f:
    loss_data = json.load(f)

# Step 2: Prepare data for plotting
iterations = list(map(int, loss_data.keys()))  # Convert keys to integers (iterations)
loss_values = list(loss_data.values())  # Extract loss values


# Step 3: Create the plot using Plotly
fig = go.Figure()

# Add a trace for the loss values
fig.add_trace(go.Scatter(
    x=iterations,
    y=loss_values,
    mode='lines',
    name='Loss',
    line=dict(color='blue'),
    marker=dict(symbol='circle', size=6)
))

# Step 4: Customize the plot
fig.update_layout(
    # title='Loss Decreasing Over Iterations',
    xaxis_title='Iterations',
    yaxis_title='Loss',
    template='plotly_white',
    font=dict(size=14),
    width=800,  # Width of the plot
    height=500  # Height of the plot
)

# Step 5: Display the plot
fig.show()

# Save the figure as a PDF
pio.write_image(fig, 'loss_household_generation.pdf', format='pdf')
```
